```python
%matplotlib notebook
import matplotlib.pyplot as plt
```

TEC data from Mahali <br>
MIT led NSF project studying the Earth’s ionosphere with GPS<br>
http://mahali.mit.edu/

```python
from skdaccess.framework.param_class import *
from skdaccess.geo.mahali.tec.data_fetcher import DataFetcher
```

```python
mahalidf = DataFetcher([AutoList(['mh03'])],start_date='2015-10-07', end_date='2015-10-07')
```

```python
dw = mahalidf.output()
```

```python
label, data = next(dw.getIterator())
```

```python
plt.plot(data.vertical_tec[data.elevation > 20.0],'.');
plt.title(label);
plt.ylabel('TEC');
plt.xticks(rotation=15);
```

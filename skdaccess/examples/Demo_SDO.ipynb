```python
%matplotlib notebook
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
```

Access data from the Solar Dynamics Observatory <br>
https://sdo.gsfc.nasa.gov/


```python
from skdaccess.framework.param_class import *
from skdaccess.astro.sdo import DataFetcher as SDODF
```

```python
al_urls = AutoList(['https://sdo.gsfc.nasa.gov/assets/img/browse/2017/08/28/20170828_000000_256_HMIB.jpg',
                    'http://jsoc2.stanford.edu/data/aia/synoptic/2013/04/04/H0500/AIA20130404_0500_0335.fits'])
```

```python
df = SDODF([al_urls])
```

```python
dw = df.output()
```

```python
for label, data in dw.getIterator():
    plt.figure()
    plt.axis('off')
    if label[-3:] == 'jpg':
        plt.imshow(data,cmap='gray')
    else:
        plt.imshow(data,cmap='gray',vmin=-5,vmax=20)
    plt.title(label,fontsize=8)
```
